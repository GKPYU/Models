```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt


import json
  
# Opening JSON file
f = open('2023_02_07-220832_preds.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
label=[]
y_pred_proba=[]
# Iterating through the json
# list
for i in range(len(data)):
    value=data[i]['val']
    predict=data[i]['pred']
    if data[i]['split']=='test':
        z = 1/(1 + np.exp(-predict))
        label.append(value)
        y_pred_proba.append(predict)
    
  
# Closing file
f.close()
fpr, tpr, _ = metrics.roc_curve(label,  y_pred_proba)
auc = metrics.roc_auc_score(label, y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```
